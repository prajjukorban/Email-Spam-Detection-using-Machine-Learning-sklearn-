```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("combined_data.csv")
```

```python
df
```

```python
y = df['label']
```

```python
X = df["text"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
import joblib
joblib.dump(vectorizer,"vectorizer.joblib")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train_vec,y_train)
```

```python
model.predict(X_test_vec)
```

```python
model.score(X_test_vec,y_test)
```

```python
pred = input("Enter your email: ")
pred_vec = vectorizer.transform([pred])  # Always use transform, and wrap in list
prediction = model.predict(pred_vec)

print("Result:","ðŸ›‘Spam" if prediction[0] == 1 else "âœ…Not Spam")
```
